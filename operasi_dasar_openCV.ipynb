```python
import cv2 as cv
from google.colab.patches import cv2_imshow # kalau pakai cv.imshow = crash

img = cv.imread("/gdrive/MyDrive/ASEMESTER_6/IMAGE_PROCESSING/nike.jpg")
x = img

# print(img.shape)  # dimensi gambar
# print(img.size)   # ukuran gambar
# print(img.dtype)  # tipe gambar (uint8 / integer)
# print(type(img))  # numpy.ndarry

cv2_imshow(img)
```

```python
# Menyimpan gambar
x_gray = cv.cvtColor(x, cv.COLOR_BGR2GRAY)
cv2_imshow(x_gray)
cv.imwrite("/gdrive/MyDrive/ASEMESTER_6/IMAGE_PROCESSING/x_gray.jpg", x_gray)
```

```python
# Mengubah ukuran gambar
import time

start1 = time.time()
x_res = cv.resize(x, (0,0), fx=1.5, fy=1.5, interpolation=cv.INTER_AREA)
t1 = time.time() - start1 

start2 = time.time()
x_res = cv.resize(x, (0,0), fx=1.5, fy=1.5, interpolation=cv.INTER_CUBIC)
t2 = time.time() - start2 

start3 = time.time()
x_res = cv.resize(x, (0,0), fx=1.5, fy=1.5, interpolation=cv.INTER_LINEAR)
t3 = time.time() - start3 
# cv2_imshow(x_res)

print("Area = ", t1)
print("Cubic = ", t2)
print("Linear = ", t3)
```

```python
# Memisahkan kanal
b, g, r = cv.split(x)

# cv2_imshow(x)
# cv2_imshow(b)
# cv2_imshow(g)
# cv2_imshow(r)

merge = cv.merge([b, g, r])

cv2_imshow(merge)
```

```python
# Keluaran kode ini berupa baris-kolom nilai pixel
gray = cv.cvtColor(x, cv.COLOR_BGR2GRAY)
for px in gray:
  txt = ''
  for pic in px:
    txt +='\t'+str(pic)
  print (txt)
```

```python
# Mengubah array ke data frame ke CSV
import pandas as pd

DF = pd.DataFrame(gray)
DF.to_csv("/gdrive/MyDrive/ASEMESTER_6/IMAGE_PROCESSING/nike.csv")
```
