## **DFT Manual**

```python
import numpy as np

def dft (data):
  N = len(data)
  x = np.zeros(N, dtype="complex")
  for i in range(N):
    x_total = 0
    for k in range(N):
      x_aksen = data[k] * np.exp(-1j*2*np.pi*i*k/N)
      x_total = x_total + x_aksen
    x[i] = x_total
  return x

x = [2, 3, 4, 5]
transformasi = dft(x)
magnitudo = np.abs(dft(x))
```

## **FFT dengan Numpy** 

```python
import numpy as np

# Data yang mau ditransformasi
fx = np.array([2, 3, 4, 5])
fx = np.float32(fx)

# FFT
Fu = np.fft.fft(fx)
fshift = np.fft.fftshift(Fu)

# Inverse FFT
f_ishift = np.fft.ifftshift(fshift)
img_back = np.fft.ifft(f_ishift)
img_back = np.real(img_back)

print("fx = ")
print(fx)

print("\nFu")
print(Fu)

print("\nDFT shift / geser DC components")

print("\ninvers DFT = ")
print(img_back)
```

## **DFT dengan CV2**

```python
import cv2 as cv
import numpy as np

# Data yang mau ditransformasi
fx = np.array([[1, 2, 3],
               [3, 4, 5],
               [2, 3, 4]])
fx = np.float32(fx)

# FFT
dft = cv.dft(fx, flags=cv.DFT_COMPLEX_OUTPUT)

# Magnitudo
mag = cv.magnitude(dft[:, :, 0], dft[:, :, 1])

# IFFT
idft = cv.idft(dft, flags=cv.DFT_SCALE)

# Hasil
result = cv.magnitude(idft[:, :, 0], idft[:, :, 1])

print("fx = ")
print(fx)

print("\nFu")
print(dft)

print("\ninvers DFT = ")
print(idft)

print("\nMagnitude = ")
print(result)
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Citra yang ingin diolah
fx = cv.imread("/gdrive/MyDrive/ASEMESTER_6/IMAGE_PROCESSING/hand.png", 0)
fx = np.float32(fx)

# FFT dan pergeseran nol terpusat
dft = cv.dft(fx, flags=cv.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)

# Fase spektrum hasil FFT
fase = cv.phase(dft_shift[:, :, 0], dft_shift[:, :, 1],
                angleInDegrees=True)

# Magnitudo spektrum
mag = cv.magnitude(dft_shift[:, :, 0], dft_shift[:, :, 0])

# Menormalisasi spektrum dalam rentang 0-255
nor = cv.normalize(mag, None, 0, 255, cv.NORM_MINMAX)

# Konversi magnitudo ternormalisasi ke tipe data UINT8
nor_uint8 = nor.astype(np.uint8)

# Spektrum magnitudo skala logaritmik
mag_log = np.zeros(mag.shape, dtype=mag.dtype)
cv.normalize(cv.log(1 + mag), mag_log, 0, 255, cv.NORM_MINMAX)
mag_log_uint8 = mag_log.astype(np.uint8)

plt.subplot(221)
plt.imshow(fx, cmap="gray")

plt.subplot(222)
plt.imshow(nor_uint8, cmap="gray")

plt.subplot(223)
plt.imshow(mag_log_uint8, cmap="gray")

plt.subplot(224)
plt.imshow(fase, cmap="gray")

plt.tight_layout()
plt.show()
```
